# 矩阵中的路径
## 题目描述
请设计一个函数，用来判断在一个矩阵中是否存在一条包含某字符串所有字符的路径。路径可以从矩阵中的任意一个格子开始，每一步可以在矩阵中向左，向右，向上，向下移动一个格子。如果一条路径经过了矩阵中的某一个格子，则该路径不能再进入该格子。 例如 a b c e s f c s a d e e 矩阵中包含一条字符串"bcced"的路径，但是矩阵中不包含"abcb"路径，因为字符串的第一个字符b占据了矩阵中的第一行第二个格子之后，路径不能再次进入该格子。   
```
a b c e   
s f c s   
a d e e   
```

## 吐槽
题目有漏洞啊哥哥！矩阵中存在包含字符串"bcced"所有字符的路径为"abcceda"，那么请问矩阵中是否存在包含字符串"bccd"所有字符的路径？
很明显存在啊，"abcceda"就满足条件！题目只说路径包含字符，又没说路径不能包含其他字符！甚至，打乱顺序也可以啊！遍历"a b c e s f c s a d e e"，我就问这个路径包不包含所有字符！

````python
# 本解法默认路径不能包含其他字符，不可以打乱顺序
# -*- coding:utf-8 -*-
class Solution:
    def __init__(self):
        self.matrix = []
        self.boolmatrix = []
        self.rows = 0
        self.cols = 0
        self.path = ''
        
    def hasPath(self, matrix, rows, cols, path):
        # write code here
        self.matrix = matrix
        self.boolmatrix = [[0 for i in range(cols)] for j in range(rows)]
        self.rows = rows
        self.cols = cols
        self.path = path
        
        # print(self.boolmatrix)
        
        starts = self.findStart(matrix, rows, cols, path)
        
        if len(path) == 1 and len(starts)!= 0:
            return True
        
        for start in starts:
            flag = self.oneStep(start[0],start[1],1)
            if flag == True:
                return True
        return False
        
                    
    def findStart(self, matrix, rows, cols, path):
        starts = []
        for i in range(rows):
            for j in range(cols):
                if matrix[i][j] == path[0]:
                    starts.append([i,j])
        return starts
                    
    
    def oneStep(self,row,col,nextch):
        self.boolmatrix[row][col] = 1
        
        if row > 0 and self.boolmatrix[row-1][col] == 0 and self.matrix[row-1][col] == self.path[nextch]:
            self.boolmatrix[row-1][col] = 1
            flag = True
            if nextch <= len(self.path)-1:
                next_flag = self.oneStep(row-1,col,nextch+1)
                flag = flag and next_flag
                return flag
            else:
                return flag
            
        if col > 0 and self.boolmatrix[row][col-1] == 0 and self.matrix[row][col-1] == self.path[nextch]:
            self.boolmatrix[row][col-1] = 1
            flag = True
            if nextch < len(self.path)-1:
                next_flag = self.oneStep(row,col-1,nextch+1)
                flag = flag and next_flag
                return flag
            else:
                return flag
            
        if row < self.rows - 1 and self.boolmatrix[row+1][col] == 0 and self.matrix[row+1][col] == self.path[nextch]:
            self.boolmatrix[row+1][col] = 1
            flag = True
            if nextch < len(self.path)-1:
                next_flag = self.oneStep(row+1,col,nextch+1)
                flag = flag and next_flag
                return flag
            else:
                return flag
            
        if col < self.cols - 1 and self.boolmatrix[row][col+1] == 0 and self.matrix[row][col+1] == self.path[nextch]:
            self.boolmatrix[row][col+1] = 1
            flag = True
            if nextch < len(self.path)-1:
                next_flag = self.oneStep(row,col+1,nextch+1)
                flag = flag and next_flag
                return flag
            else:
                return flag

        return False
                    
        
def test():
    matrix = [['a','b','c','e'],['s','f','c','s'],['a','d','e','d']]
    rows = len(matrix)
    cols = len(matrix[0])
    path = 'bcced'   
#     print(matrix[rows-1][cols-1])
    
    solution = Solution()
#     starts = solution.findStart(matrix,rows,cols,path)
#     print(starts)
    result = solution.hasPath(matrix,rows,cols,path)
    print(result)
    
test()
````

````python
# 标准答案
# -*- coding:utf-8 -*-
class Solution:
    def hasPath(self, matrix, rows, cols, path):
        # write code here
        for i in range(rows):
            for j in range(cols):
                if matrix[i*cols+j] == path[0]:
                    if self.find(list(matrix),rows,cols,path[1:],i,j):
                        return True
        return False
    def find(self,matrix,rows,cols,path,i,j):
        if not path:
            return True
        matrix[i*cols+j]='0'
        if j+1<cols and matrix[i*cols+j+1]==path[0]:
            return self.find(matrix,rows,cols,path[1:],i,j+1)
        elif j-1>=0 and matrix[i*cols+j-1]==path[0]:
            return self.find(matrix,rows,cols,path[1:],i,j-1)
        elif i+1<rows and matrix[(i+1)*cols+j]==path[0]:
            return self.find(matrix,rows,cols,path[1:],i+1,j)
        elif i-1>=0 and matrix[(i-1)*cols+j]==path[0]:
            return self.find(matrix,rows,cols,path[1:],i-1,j)
        else:
            return False
        
def test():
    matrix = 'ABCEHJIGSFCSLOPQADEEMNOEADIDEJFMVCEIFGGS'
    rows = 5
    cols = 8
    path = 'SGGFIECVAASABCEHJIGQEM'

    solution = Solution()
    result = solution.hasPath(matrix, rows, cols, path)
    print(result)

test()
````
